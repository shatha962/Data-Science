```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
fpath = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/titanic_dirty_cleaned.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
# Plot the counts for the number of passengers who Survived vs. those that died.
val_counts  = df['Survived'].value_counts()
val_counts
```

```python
ax = sns.countplot(data=df, x='Survived')
ax.set_title('# of Passengers who survived Vs. died');
```

```python
# Plot a histogram of the distribution of Age.
ax = sns.histplot(data=df, x = 'Age');
ax.set_title('Distribution of age');
```

```python
# Plot a boxplot for the Fare column.
df['Fare'].value_counts()
```

```python
ax =sns.boxplot(data=df, x='Fare')
ax.set_title('Boxplt for Fare');
```
