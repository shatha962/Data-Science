```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Read in data using direct url
athlete_url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vTtqWeFFAV3GA13M4eXdzMkAKAxIQ7fek4kv16wntAI_2QfYzm_BxdW6HVblVHy0VFRwsA47xFn8--6/pub?output=csv"
df = pd.read_csv(athlete_url)
df.head()
```

```python
# Get correlations
corr = df.corr(numeric_only=True)
corr.round(2)
```

```python
# Make a heatmap of correlations
sns.heatmap(corr, cmap='Blues', annot=True);
```

```python
fpath="/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/ames-housing-cleaned-eda.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
# Get correlations
corr = df.corr(numeric_only = True)
corr.round(2)
```

```python
# Make heatmap of correlations
sns.heatmap(corr, cmap = 'coolwarm', annot = True);
```

```python
# Create a fig, ax with a figsize argument
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(corr, cmap = 'coolwarm', annot = True, ax=ax);
```
